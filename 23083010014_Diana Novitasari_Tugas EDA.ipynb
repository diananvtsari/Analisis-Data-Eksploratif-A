Deskripsi Kasus
Data yang Anda berikan berisi informasi medis pasien dengan beberapa fitur yang dapat digunakan untuk menganalisis kondisi kesehatan jantung mereka. Berikut penjelasan dari setiap kolom dalam data ini:
1. Age : Usia pasien dalam tahun
2. Sex : Jenis kelamin pasien, dengan kode:
   - M : Male (Laki-laki)
   - F : Female (Perempuan)
3. ChestPainType : Jenis nyeri dada yang dialami pasien, dengan kode:
   - ATA: Atypical Angina (Angina atipikal)
   - NAP: Non-Anginal Pain (Nyeri bukan angina)
   - ASY : Asymptomatic (Tidak bergejala)
   - TA: Typical Angina (Angina tipikal)
4. RestingBP : Tekanan darah sistolik pasien saat istirahat, diukur dalam mm Hg (millimeter merkuri).
5. Cholesterol : Tingkat kolesterol serum dalam mg/dL (milligram per desiliter).
6. FastingBS : Kadar gula darah puasa, dengan nilai:
   -  1 : Gula darah puasa â‰¥ 120 mg/dL
   -  0 : Gula darah puasa < 120 mg/dL
7. RestingECG : Hasil elektrokardiografi istirahat, dengan kode:
   - Normal : Hasil normal
   -  ST : Ada kelainan segmen ST atau T (misalnya, depresi atau elevasi ST)
   -  LVH : Hipertrofi ventrikel kiri yang menunjukkan kelainan lain
8. MaxHR : Denyut jantung maksimum yang dicapai pasien selama uji stres fisik
9. ExerciseAngina : Angina yang disebabkan oleh aktivitas fisik, dengan kode:
   -  Y : Yes (Ya, ada angina saat berolahraga)
   -  N : No (Tidak ada angina saat berolahraga)
10. Oldpeak : Depresi ST yang diinduksi oleh latihan relatif terhadap istirahat (dalam mm).

11. ST_Slope : Kemiringan segmen ST saat latihan, dengan kode:
   - Up : Kemiringan naik
   -  Flat : Kemiringan datar
   - Down : Kemiringan turun
12. HeartDisease : Indikasi apakah pasien memiliki penyakit jantung atau tidak, dengan kode:
   -  1 : Menderita penyakit jantung
   -  0 : Tidak menderita penyakit jantung

# Scatter Plot ini digunakan untuk melihat korelasi antara tekanan darah dan kadar kolesterol

```python
plt.scatter(data['RestingBP'], data['Cholesterol'])
plt.xlabel('RestingBP')
plt.ylabel('Cholesterol')
plt.title('Relationship between RestingBP and Cholesterol')
plt.show()
```

# Bar Chart ini digunakan ntuk membandingkan frekuensi masing-masing jenis nyeri dada pada pasien dengan atau tanpa penyakit jantung

```python
heart_disease_yes = data[data['HeartDisease'] == 1]
heart_disease_no = data[data['HeartDisease'] == 0]

chest_pain_types = data['ChestPainType'].unique()
yes_counts = [len(heart_disease_yes[heart_disease_yes['ChestPainType'] == chest_pain_type]) for chest_pain_type in chest_pain_types]
no_counts = [len(heart_disease_no[heart_disease_no['ChestPainType'] == chest_pain_type]) for chest_pain_type in chest_pain_types]

plt.bar(chest_pain_types, yes_counts, label='With Heart Disease')
plt.bar(chest_pain_types, no_counts, bottom=yes_counts, label='Without Heart Disease')
plt.xlabel('ChestPainType')
plt.ylabel('Frequency')
plt.title('Frequency of ChestPainType by HeartDisease')
plt.legend()
plt.show()
```

# Box Plot ini digunakan untuk mengevaluasi seberapa besar pengaruh latihan pada detak jantung maksimal dan korelasinya dengan penyakit jantung

```python
plt.boxplot([data[data['HeartDisease'] == 0]['MaxHR'], data[data['HeartDisease'] == 1]['MaxHR']], labels=['Without Heart Disease', 'With Heart Disease'])
plt.xlabel('HeartDisease')
plt.ylabel('MaxHR')
plt.title('Box Plot of MaxHR by HeartDisease')
plt.show()
```

# Stacked Bar Chart ini digunakan untuk melihat pengaruh latihan angina terhadap kondisi jantung berdasarkan variabel HeartDisease.

```python
exercise_angina_counts = data['ExerciseAngina'].value_counts()
heart_disease_yes = data[data['HeartDisease'] == 1]
heart_disease_no = data[data['HeartDisease'] == 0]

yes_counts = [len(heart_disease_yes[heart_disease_yes['ExerciseAngina'] == exercise_angina]) for exercise_angina in exercise_angina_counts.index]
no_counts = [len(heart_disease_no[heart_disease_no['ExerciseAngina'] == exercise_angina]) for exercise_angina in exercise_angina_counts.index]

plt.bar(exercise_angina_counts.index, yes_counts, label='With Heart Disease')
plt.bar(exercise_angina_counts.index, no_counts, bottom=yes_counts, label='Without Heart Disease')
plt.xlabel('ExerciseAngina')
plt.ylabel('Frequency')
plt.title('Stacked Bar Chart of ExerciseAngina by HeartDisease')
plt.legend()
plt.show()
```

# Clustered Bar Chart digunakan untuk melihat bagaimana variabel-variabel kategori ini terkait dengan penyakit jantung


```python
categories = ['Sex', 'ChestPainType', 'RestingECG', 'ST_Slope']
heart_disease_yes = data[data['HeartDisease'] == 1]
heart_disease_no = data[data['HeartDisease'] == 0]

for category in categories:
    yes_counts = heart_disease_yes[category].value_counts()
    no_counts = heart_disease_no[category].value_counts()
    
    plt.bar(yes_counts.index, yes_counts.values, label='With Heart Disease')
    plt.bar(no_counts.index, no_counts.values, bottom=yes_counts.values, label='Without Heart Disease')
    plt.xlabel(category)
    plt.ylabel('Frequency')
    plt.title(f'Clustered Bar Chart of {category} by HeartDisease')
    plt.legend()
    plt.show()
```

# Clustered Bar Chart digunakan untuk melihat bagaimana variabel-variabel kategori ini terkait dengan penyakit jantung

```python
numeric_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
numeric_data = data[numeric_cols]

corr_matrix = numeric_data.corr()

plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

# Pie Chart untuk melihat presentase pasien memiliki penyakit jantung atau tidak

```python
labels = ['No Heart Disease', 'Heart Disease']
sizes = [len(data[data['HeartDisease'] == 0]), len(data[data['HeartDisease'] == 1])]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0) 

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
ax.axis('equal')  

plt.title('Distribution of Heart Disease')
plt.show()
```
